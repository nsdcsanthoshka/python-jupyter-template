### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the line chart (Subplot 1)
x_line = np.arange(0, 11)
y_line = x_line**2

# Data for the bar chart (Subplot 2)
categories_bar = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]
colors_bar = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

# Data for the scatter plot (Subplot 3)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for the pie chart (Subplot 4)
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create the figure and the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # figsize for better spacing

# Plot the line chart in the first subplot (row 0, column 0)
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')

# Plot the bar chart in the second subplot (row 0, column 1)
axs[0, 1].bar(categories_bar, values_bar, color=colors_bar)
axs[0, 1].set_title('Bar Chart')

# Plot the scatter plot in the third subplot (row 1, column 0)
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Plot the pie chart in the fourth subplot (row 1, column 1)
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].axis('equal')
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```
